```python
dataset = load_iris(return_X_y=True, as_frame=False)
```

```python

X = dataset[0]

Y = dataset[1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=8,test_size=0.3)

```

```python


# convert integers to dummy variables (i.e. one hot encoded)
dummy_y_test = np_utils.to_categorical(y_test)
dummy_y_test
```

```python

# define baseline model
def baseline_model(optm,act_func):
    
    # create model
    model = Sequential()
    
    model.add(Dense(8, input_dim=4, activation=act_func))
    
    model.add(Dense(3, activation='softmax'))
    
    
    # Compile model
    model.compile(loss='categorical_crossentropy', 
                  optimizer=optm, 
                  metrics=['accuracy'])
    
    
    return model
```

```python
m1 = baseline_model("sgd","relu")

m3 = baseline_model("adam","relu")

m1.layers[0].activation
```

```python
m1.fit(X_train,
       dummy_y_train, 
       epochs=100,
       batch_size=10,
      verbose=True)
```
