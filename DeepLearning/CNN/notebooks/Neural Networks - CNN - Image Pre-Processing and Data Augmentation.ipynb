```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

start_point = (5, 5) 
  
end_point = (220, 220) 
  
# Blue color in BGR 
color = (255, 0, 0) 
thickness = 2

image = cv2.rectangle(image, start_point, end_point, color, thickness) 

cv2.imshow('boundary image', image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
for i in range(40):
    start_point = (5+i*2, 5) 
  
    end_point = (220+i*2, 220) 
  
    # Blue color in BGR 
    color = (2+i, i, i*2) 
    thickness = 2

    image = cv2.rectangle(image, start_point, end_point, color, thickness) 

    cv2.imshow('boundary image', image)

    cv2.waitKey(0)

    cv2.destroyAllWindows()
```
