```python
#RANDOM US

#the class labels, 0 , and 1 , are considered as Negative and Positive 


# selecting the indices of the non-fraud cases
fraud_indices = df[df.Class == 1].index
nonfraud_indices = df[df.Class == 0].index

# From all non-fraud cases, randomly select cases equal to number of fraud cases
random_nonfraud_indices = np.random.choice(nonfraud_indices, frauds, replace = False)


random_nonfraud_indices = np.array(random_nonfraud_indices)

# Concat or attache the 2 indices
under_sample_indices = np.concatenate([fraud_indices,random_nonfraud_indices])

# Under sample dataset
under_sample_data = df.iloc[under_sample_indices,:]
under_sample_data
```

```python


# Now split X, y variables from the under sample data
#X_undersample = under_sample_data.drop(["Class"],axis=1)

X_undersample = under_sample_data.loc[:, under_sample_data.columns != 'Class']
y_undersample = under_sample_data.loc[:, under_sample_data.columns == 'Class']


pd.value_counts(under_sample_data['Class'])

```

### Note that in binary classification, recall of the positive class is also known as “sensitivity”; recall of the negative class is “specificity”.

### Sensitivity/recall – how good a test is at detecting the positives. A test can cheat and maximize this by always returning “positive”.

### Specificity – how good a test is at avoiding false alarms. A test can cheat and maximize this by always returning “negative”
