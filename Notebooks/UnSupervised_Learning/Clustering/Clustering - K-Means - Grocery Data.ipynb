```python
data=pd.read_csv("Wholesale customers data.csv")
data.head()
```

```python
#plot with clustering

import matplotlib.pyplot as plt

plt.scatter(data.Fresh, data.Frozen)
plt.show()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_scaled = scaler.fit_transform(data)

# statistics of scaled data
pd.DataFrame(data_scaled).describe()
```

```python

kmeans = KMeans(n_clusters=2, init='k-means++', verbose=True, random_state =100, max_iters=300)


#??kmeans
```

```python

# fitting the k means algorithm on scaled data

kmeans.fit(data_scaled)
```

```python
kmeans.inertia_ #WSS

kmeans.cluster_centers_
```

```python
#SSE is Intertia or WSS

SSE = []

for cluster in range(2,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

    
#min clusters >=2
#max clusters <=number of observations i.e. N [Ideally <= N-1 ]
```

```python
#just to check if for loop working fine

for i in range(1,20):
    print(i)
```

```python

# converting the results into a dataframe and plotting them

frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})


plt.figure(figsize=(12,6))


plt.plot(frame['Cluster'], frame['SSE'], marker='o')

plt.xlabel('Number of clusters')

##Inertia is the error
plt.ylabel('Inertia')

```

```python
# k means using 5 clusters and k-means++ initialization

kmeans = KMeans(n_jobs = -1, n_clusters = 5, init='k-means++')

kmeans.fit(data_scaled)

kmeans.cluster_centers_
```
